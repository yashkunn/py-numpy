````python
# Python code
def unknown_signature(lst, mask, new_value):
    for i in range(len(lst)):
        if mask[i]:
            lst[i] = new_value

lst = [1, 2, 3, 4, 5]
mask = [True, False, True, False, True]
new_value = 0
unknown_signature(lst, mask, new_value)
lst
````

````python
def unknown_signature_numpy(lst, mask, new_value):
    lst = np.array(lst)
    mask = np.array(mask)
    
    lst[mask] = new_value
    return lst

lst = [1, 2, 3, 4, 5]
mask = [True, False, True, False, True]
new_value = 0
unknown_signature_numpy(lst, mask, new_value)
````

### Task 7.2

````python
# Python code
def unknown_signature(lst1, lst2):
    bitwise_complement = [~x for x in lst1]
    bitwise_and = [x & y for x, y in zip(lst1, lst2)]
    bitwise_or = [x | y for x, y in zip(lst1, lst2)]
    return bitwise_complement, bitwise_and, bitwise_or

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]
result_bitwise_complement, result_bitwise_and, result_bitwise_or = unknown_signature(list1, list2)
result_bitwise_complement, result_bitwise_and, result_bitwise_or
````

````python
def unknown_signature_numpy(lst1, lst2):
    arr1 = np.array(lst1)
    arr2 = np.array(lst2)
    
    bitwise_complement = np.bitwise_not(arr1)
    
    bitwise_and = np.bitwise_and(arr1, arr2)
    
    bitwise_or = np.bitwise_or(arr1, arr2)
    
    return bitwise_complement, bitwise_and, bitwise_or

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]
unknown_signature_numpy(list1, list2)
````

````python
# Python code
def unknown_signature(lst):
    result = []
    for num in lst:
        if num % 2 == 0 and (num % 3 == 0 or num > 10):
            modified_num = (num ^ 7) + 2
            result.append(modified_num)
    return result

my_list = [6, 9, 12, 14, 17, 20]
result_python_style = unknown_signature(my_list)
result_python_style
````

````python
def unknown_signature_numpy(lst):
    arr = np.array(lst)
    mask = (arr % 2 == 0) & ((arr % 3 == 0) | (arr > 10))
    modified_arr = (arr[mask] ^ 7) + 2
    return modified_arr

my_list = [6, 9, 12, 14, 17, 20]
unknown_signature_numpy(my_list)
````
